 # ***Data Cleaning***

```python
#I started changing the column names to more understandable ones
df_sales_predictions.columns
```

```python
df_sales_predictions= df_sales_predictions.rename({'Item_Identifier':'Item_Identifier', 'Item_Weight':'Item_Weight', 'Item_Fat_Content':'Item_Fat_Content',
      'Item_Visibility':'Item_Visibility','Item_Type':'Item_Type',
       'Item_MRP':'Retail_Price', 'Outlet_Identifier':'Outlet_Identifier',
       'Outlet_Establishment_Year':'Establishment_Year', 'Outlet_Size':'Outlet_Size', 'Outlet_Location_Type':'Outlet_Location_Type',
       'Outlet_Type':'Outlet_Type','Item_Outlet_Sales':'Item_Outlet_Sales'},axis=1)
df_sales_predictions
```

```python
#I noticed that the word "Tier" is repeated throughout the column with the number, and we could do without it, so I'll replace it with '' and change the data type for the column.
df_sales_predictions['Outlet_Location_Type']=df_sales_predictions['Outlet_Location_Type'].str.replace('Tier','')
df_sales_predictions['Outlet_Location_Type'].value_counts()#to confirm the change
```

```python
df_sales_predictions['Outlet_Location_Type']=df_sales_predictions['Outlet_Location_Type'].astype(float)# to change the datatype of 'Outlet_Location_Type' column to float

df_sales_predictions['Outlet_Location_Type'].dtype # to confirm the datatype
```

# **Missing Values**

```python
#I want to check the null values in which column they are present and how many
df_sales_predictions.isna().sum()
```

```python
df_sales_predictions.isna().sum()/len(df_sales_predictions)*100
```

```python
# Visualizition the missing values using the missingno package's matrix plot.
import missingno as msno
msno.matrix(df_sales_predictions);
```

```python
df_sales_predictions['Item_Weight'].value_counts()
```

```python
#df_sales_predictions.loc[['Item_Weight','Item_Identifier','Item_Type'	]]
null_garage_type = df_sales_predictions['Item_Weight'].isna()#

null_Item = df_sales_predictions[null_garage_type]
null_Item
```

```python
null_Item.sort_values('Item_Identifier')
```

```python
#df_sales_predictions['Item_Weight'].mean()
#df_sales_predictions.groupby('Item_Weight')
groubBy_Identifier = df_sales_predictions.groupby('Item_Identifier')['Item_Weight'].mean()
groubBy_Identifier
```

```python
#df_sales_predictions = df_sales_predictions.set_index("Item_Identifier")
groubBy_Identifier['DRA24']
```

```python
for key,value in df_sales_predictions['Item_Weight'].items():


   value_Item_Identifier= df_sales_predictions['Item_Identifier'][key]
   from_groubBy_Identifier = groubBy_Identifier[value_Item_Identifier]
   df_sales_predictions['Item_Weight'][key]= from_groubBy_Identifier

```

```python
null_Item
```

```python
df_sales_predictions['Item_Weight'].isna().sum()
```

```python
#df_sales_predictions['Alley'] = df['Alley'].fillna("MISSING")
df_sales_predictions['Item_Weight']=df_sales_predictions['Item_Weight'].fillna(df_sales_predictions['Item_Weight'].mean())
```
